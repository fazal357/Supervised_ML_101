```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
house_p = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
print(house_p)
```

```python
house_p.head(5)
```

```python
house_p['price']=housing.target
```

```python
corr_mat= house_p.corr()
```

```python
print(corr_mat)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr_mat, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
X = house_p.drop(['price'], axis=1)
Y = house_p['price']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model= XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
train_pred=model.predict(X_train)
```

```python
score1= metrics.r2_score(Y_train, train_pred)
score2= metrics.mean_absolute_error(Y_train, train_pred)

print("R2 Score Prediction: ", score1)
print("MAE Score Prediction: ", score2)
```

```python
plt.scatter(Y_train, train_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Price Vs Predicted Price")
plt.show()
```

```python
test_pred= model.predict(X_test)
```

```python
score3 = metrics.r2_score(Y_test, test_pred)
score4 = metrics.mean_absolute_error(Y_test, test_pred)

print("R2 Score Prediction: ", score3)
print("MAE Score Prediction: ", score4)
```

```python
plt.scatter(Y_test, test_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Price Vs Predicted Price")
plt.show()
```
